# Trends

1. Temperature gets hotter as the latitude approaches equator 
2. Humidity values peak as the latitude approaches equator
3. Wind speed decreases as the latitude approaches equator


```python
# Dependencies
import csv
import matplotlib.pyplot as plt
import requests
import pandas as pd
from config import api_key
from citipy import citipy
import numpy as np
from random import uniform
```

```python
# Latitude range from -90째 at the South Pole to +90째 at the North Pole. 
# Longitude range from -180째 when traveling west to 180째 when traveling east.
# generate uniform distribution of lat & long using numpy's random unifom

latitude = np.random.uniform(-90, 90, 1500)
longitude = np.random.uniform(-180, 180, 1500)

# declare cities & coordinates

coordinates = []
cities = []

# get coordinates by combining lat & long

coordinates = zip(latitude, longitude)

# get nearest cities for coordinates using citipy

for i in coordinates:
    city = citipy.nearest_city(i[0], i[1]).city_name
    if city not in cities:
        cities.append(city)


#cities_subset = cities[:50]
```

```python
# Save config information.
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# Build partial query URL

query_url = f"{url}appid={api_key}&units={units}&q="
query_url

lat = []
long = []
country = []
temp = []
humidity = []
clouds = []
wind = []
date = []
city_found = []

print("Beginning Data Retrieval     ")
print("-----------------------------")


# Loop through the list of cities and perform a request for data on each and make a list of cities found in open weather

for city in cities:  
    city_url = query_url + city
    response = requests.get(query_url + city).json()
    print(city_url)
    try:
        response = requests.get(query_url + city).json()
        lat.append(response['coord']['lat'])
        long.append(response['coord']['lon'])
        country.append(response['sys']['country'])
        temp.append(response['main']['temp_max'])
        humidity.append(response['main']['humidity'])
        clouds.append(response['clouds']['all'])
        wind.append(response['wind']['speed'])
        date.append(response['dt'])
        city_found.append(response['name'])

        
    except(KeyError):
        print("Missing city in openweather... skipping.")
        
              
# Indicate that Data Loading is complete 
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")

    
```

```python
# convert to a dataframe

cities_df = pd.DataFrame({"City": city_found, "Lat": lat, "Lng": long, "Max Temp": temp, "Humidity": humidity,"Cloudiness": clouds,
                          "Wind Speed": wind,"Country": country,"Date": date})

cities_df.count()

cities_df.to_csv("cities_data.csv", index=False)
```

```python
# Build scatter plot for latitude vs temperature
plt.scatter(cities_df["Lat"], 
            cities_df["Max Temp"],
            edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8, label="Cities")

plt.title("City Latitude vs. Max Temperature")
plt.ylabel("Max Temperature (F)")
plt.xlabel("Latitude")
plt.grid(True)


# Save the figure
plt.savefig("Latitude vs Temperature (F).png")

# Show plot
plt.show()
```
